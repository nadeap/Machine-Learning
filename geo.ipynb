```python
import matplotlib.pyplot as plt
```

```python
x = ['Sidar','Andi','Dirwan','Naim','Ayu','Umiyati','Intan']
y = [400.000,100.000,150.000,400.000,300.000,300.000,200.000]
zipped = zip(y, x)
zipped = sorted(zipped, reverse=True)
y, x = zip(*zipped)

plt.bar(x,y,color='red')
plt.xlabel('Nama Penjual')
plt.ylabel('Pendapatan Per Hari')
plt.title('Hasil Pendapatan Penjual Lokawisata Baturaden')
plt.show()
```

```python
x = ['Hujan Ramai','Panas Ramai']
y = [3,4]

plt.bar(x,y,color='red')
plt.xlabel('Keramaian')
plt.ylabel('Jumlah Penjual')
plt.title('Perbandingan Keramaian Pembeli')
plt.show()
```
